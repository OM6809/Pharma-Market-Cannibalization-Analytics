**1.A) Market Dynamics and Competitive Landscape Assessment**

```python
##  Aggregating Claims Data by Year and Procedure Code ##

import matplotlib.pyplot as plt
import numpy as np

# Filter data for the years 2016 to 2018
claims_filtered_years = final_df[final_df['Year'].isin([2016, 2017, 2018])]

# Group by year and HCPCS code (product) to count claims, patients, and HCPs
claims_grouped = claims_filtered_years.groupby(['Year', 'Procedure_Code']).agg(
    total_claims=('Claim_ID', 'count'),
    unique_patients=('Patient_ID', pd.Series.nunique),
    unique_hcps=('HCP_ID', pd.Series.nunique)
).reset_index()

# Calculate total claims per year for percentage calculation
total_claims_per_year = claims_grouped.groupby('Year')['total_claims'].sum()
```

```python
# Calculate share of claims, patients, and HCPs per product per year
claims_grouped['claims_percentage'] = claims_grouped.apply(lambda x: (x['total_claims'] / total_claims_per_year[x['Year']]) * 100, axis=1)
claims_grouped['patients_percentage'] = claims_grouped.apply(lambda x: (x['unique_patients'] / claims_grouped[claims_grouped['Year'] == x['Year']]['unique_patients'].sum()) * 100, axis=1)
claims_grouped['hcps_percentage'] = claims_grouped.apply(lambda x: (x['unique_hcps'] / claims_grouped[claims_grouped['Year'] == x['Year']]['unique_hcps'].sum()) * 100, axis=1)

# Pivot data for stacked bar charts
claims_percentage_pivot = claims_grouped.pivot(index='Year', columns='Procedure_Code', values='claims_percentage')
patients_percentage_pivot = claims_grouped.pivot(index='Year', columns='Procedure_Code', values='patients_percentage')
hcps_percentage_pivot = claims_grouped.pivot(index='Year', columns='Procedure_Code', values='hcps_percentage')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display  # <-- Add this

# Apply a professional style
plt.style.use('seaborn-v0_8-whitegrid')

# Custom professional color palette
colors = sns.color_palette("husl")

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(14, 16), sharex=True)

# Function to add labels to bars
def add_labels(ax, data):
    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f%%', label_type='center', color='black', fontsize=9)

# Function to plot stacked bar chart
def stacked_bar_chart(ax, data, title, ylabel):
    data_perc = data.divide(data.sum(axis=1), axis=0) * 100
    data_perc.plot(kind='bar', stacked=True, ax=ax, color=colors, edgecolor='white')

    ax.set_ylabel(ylabel, fontsize=12)
    ax.set_title(title, fontsize=14, fontweight='bold')

    ax.tick_params(axis='x', labelrotation=25, labelsize=11)
    ax.tick_params(axis='y', labelsize=11)

    add_labels(ax, data_perc)

# Plot claims percentage
stacked_bar_chart(axes[0], claims_percentage_pivot,
                  'Trend of Claims Percentage for the Products',
                  'Claims Percentage')

# Plot unique patients
stacked_bar_chart(axes[1], patients_percentage_pivot,
                  'Trend of Unique Patients for the Products',
                  'Unique Patients Percentage')

# Plot unique HCPs
stacked_bar_chart(axes[2], hcps_percentage_pivot,
                  'Trend of Unique HCPs for Products',
                  'Unique HCPs Percentage')

# Improved legend styling
plt.legend(title='Procedure Code / Products',
           loc='upper right',
           bbox_to_anchor=(1.25, 1),
           fontsize=10,
           title_fontsize=11,
           frameon=True)

plt.tight_layout()
plt.show()
```

```python
claims_percentage_pivot.to_csv('claims_percentage_pivot.csv', index=True)
patients_percentage_pivot.to_csv('patients_percentage_pivot.csv', index=True)
hcps_percentage_pivot.to_csv('hcps_percentage_pivot.csv', index=True)

fig.savefig('stacked_bar_charts.png')
```

```python
#  Average Claims per HCP by Year and Procedure Code ##

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Filter data for the years 2016 to 2018 and the specified procedure codes
filtered_df = final_df[
    final_df['Year'].isin([2016, 2017, 2018]) &
    final_df['Procedure_Code'].isin(['J1885', 'J2250', 'J2704', 'J3010'])
]

# Group and calculate for both Patients and Claims
grouped_data = filtered_df.groupby(['Year', 'Procedure_Code'])
patients_per_year_product = grouped_data['Patient_ID'].nunique().reset_index()
claims_per_year_product = grouped_data['Claim_ID'].count().reset_index()
hcps_per_year_product = grouped_data['HCP_ID'].nunique().reset_index()
```

```python
# Merge the DataFrames on 'Year' and 'Procedure_Code'
merged_patients = pd.merge(patients_per_year_product, hcps_per_year_product, on=['Year', 'Procedure_Code'])
merged_claims = pd.merge(claims_per_year_product, hcps_per_year_product, on=['Year', 'Procedure_Code'])

# Calculate patients and claims per writer (HCP)
merged_patients['Patients_per_HCP'] = merged_patients['Patient_ID'] / merged_patients['HCP_ID']
merged_claims['Claims_per_HCP'] = merged_claims['Claim_ID'] / merged_claims['HCP_ID']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Professional palette
palette = sns.color_palette("husl")

sns.set_style("whitegrid")

# Plotting
fig, ax = plt.subplots(2, 1, figsize=(14, 14))

# Patients per HCP
sns.lineplot(
    data=merged_patients, x='Year', y='Patients_per_HCP',
    hue='Procedure_Code', marker='o', markersize=8,
    linewidth=2.5, palette=palette, ax=ax[0]
)
ax[0].set_title('Patients per Writer (HCP) per Year - By Product',
                fontsize=14, fontweight='bold')
ax[0].set_xlabel('')
ax[0].set_ylabel('Patients per HCP', fontsize=12)
ax[0].tick_params(axis='x', rotation=30)
ax[0].legend(title='Procedure Code', fontsize=10, title_fontsize=11)

# Annotations
for i in range(len(merged_patients)):
    ax[0].annotate(
        f"{merged_patients['Patients_per_HCP'][i]:.2f}",
        (merged_patients['Year'][i], merged_patients['Patients_per_HCP'][i]),
        textcoords="offset points", xytext=(0, 8),
        ha='center', fontsize=9
    )

# Claims per HCP
sns.lineplot(
    data=merged_claims, x='Year', y='Claims_per_HCP',
    hue='Procedure_Code', marker='o', markersize=8,
    linewidth=2.5, palette=palette, ax=ax[1]
)
ax[1].set_title('Claims per Writer (HCP) per Year - By Product',
                fontsize=14, fontweight='bold')
ax[1].set_xlabel('Year', fontsize=12)
ax[1].set_ylabel('Claims per HCP', fontsize=12)
ax[1].tick_params(axis='x', rotation=30)
ax[1].legend(title='Procedure Code', fontsize=10, title_fontsize=11)

# Annotations
for i in range(len(merged_claims)):
    ax[1].annotate(
        f"{merged_claims['Claims_per_HCP'][i]:.2f}",
        (merged_claims['Year'][i], merged_claims['Claims_per_HCP'][i]),
        textcoords="offset points", xytext=(0, 8),
        ha='center', fontsize=9
    )

plt.tight_layout()
plt.show()
```

```python
## Calculating Year-over-Year Change in Claims Volume for Variant Brand (J2250) by Territory

# Filter data for the variant brand (J2250) and years 2017 and 2018
variant_brand_data = territory_claims_volume[
    (territory_claims_volume['Procedure_Code'] == 'J2250') &
    (territory_claims_volume['Year'].isin([2017, 2018]))
]

# Pivot the data to have years as columns
variant_brand_pivot = variant_brand_data.pivot(
    index='Territory Name', columns='Year', values='Claims_Volume'
).reset_index()

# Calculate the year-over-year change percentage
variant_brand_pivot['YoY_Change_Percentage'] = ((variant_brand_pivot[2018] - variant_brand_pivot[2017]) / variant_brand_pivot[2017]) * 100

# Display the resulting DataFrame
variant_brand_pivot
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# COLOR PALETTE (FIXED ORDER)
# -------------------------------
pastel_palette = [
    '#8fbce6',  # pastel blue
    '#c27a7a',  # pastel maroon
    '#8fd19e',  # pastel green
    '#4d4d4d'   # soft black / charcoal
]

procedure_order = ['J1885', 'J2250', 'J3010', 'J2704']

palette_map = dict(zip(procedure_order, pastel_palette))

# -------------------------------
# 1. FILTER DATA
# -------------------------------
filtered_df = final_df[
    final_df['Year'].isin([2016, 2017, 2018]) &
    final_df['Procedure_Code'].isin(procedure_order)
]

# -------------------------------
# 2. NEW WRITERS
# -------------------------------
first_year = (
    filtered_df
    .groupby(['HCP_ID', 'Procedure_Code'])['Year']
    .min()
    .reset_index()
)

new_writers = first_year[first_year['Year'].isin([2016, 2017, 2018])]

new_writer_counts = (
    new_writers
    .groupby(['Year', 'Procedure_Code'])['HCP_ID']
    .nunique()
    .reset_index(name='New_Writer_Count')
)

# -------------------------------
# 3. CONTINUING WRITERS
# -------------------------------
total_writers = (
    filtered_df
    .groupby(['Year', 'Procedure_Code'])['HCP_ID']
    .nunique()
    .reset_index(name='Total_Writer_Count')
)

writers_combined = pd.merge(
    total_writers,
    new_writer_counts,
    on=['Year', 'Procedure_Code'],
    how='left'
).fillna(0)

writers_combined['Continuing_Writer_Count'] = (
    writers_combined['Total_Writer_Count'] -
    writers_combined['New_Writer_Count']
)

continuing_df = writers_combined[writers_combined['Year'].isin([2017, 2018])]

# -------------------------------
# 4. CLUSTERED BAR CHART  NEW WRITERS
# -------------------------------
plt.figure(figsize=(10, 6))
ax1 = sns.barplot(
    data=new_writer_counts,
    x='Year',
    y='New_Writer_Count',
    hue='Procedure_Code',
    palette=palette_map
)

plt.title('New Writers per Brand per Year (20162018)')
plt.xlabel('Year')
plt.ylabel('Number of New Writers')
plt.legend(title='Procedure Code')

for p in ax1.patches:
    ax1.annotate(
        f"{int(p.get_height())}",
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='bottom', xytext=(0, 6),
        textcoords='offset points'
    )

plt.show()

# -------------------------------
# 5. LINE CHART  CONTINUING WRITERS
# -------------------------------
plt.figure(figsize=(10, 6))
ax2 = sns.lineplot(
    data=continuing_df,
    x='Year',
    y='Continuing_Writer_Count',
    hue='Procedure_Code',
    palette=palette_map,
    marker='o'
)

plt.title('Continuing Writers per Brand (20172018)')
plt.xlabel('Year')
plt.ylabel('Number of Continuing Writers')
plt.xticks([2017, 2018])
plt.legend(title='Procedure Code')

for line in ax2.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax2.text(x, y, f"{int(y)}", ha='center', va='bottom')

plt.show()

```

**Observations:**

Fluctuating New Writer Trends:

1)- Varied annual trends in the number of new HCPs prescribing each brand, with some brands seeing increases and others decreases.

2)- Overall Decline in Continuing Writers: Most brands are experiencing a decline in the number of continuing writers, particularly noticeable between 2017 and 2018.

**Insights:**

1)- New Writer Acquisition: Gaining new HCPs who prescribe the product for the first time is critical for market growth and expansion.

2)- Writer Retention Challenges: Keeping existing writers engaged is key to maintaining market share and ensuring sustained product performance.

3)- Competitive Dynamics: Fluctuations in new and continuing writers may be indicative of the competitive forces at play within the market.

b. Offer incentives: Introduce programs that motivate first-time prescriptions and facilitate product trial.

c. Expand outreach: Broaden efforts to include HCPs from new territories or specialties that have not been heavily targeted.
